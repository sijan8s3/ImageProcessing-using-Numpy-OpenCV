```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#cv2 is open CV module
#cv is for computer vision
# but cv2 uses BGR instead of RGB
```

```python
img= cv2.imread("steven.jpg")
```

```python
plt.imshow(img)
#so the color in image is reverse
```

```python
#converting back to RGB
rgb= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb)
```

```python
        ### FACE ###
```

```python
#for face we use cascase classifier 
```

```python
classifier= cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
faces= classifier.detectMultiScale(img)
```

```python
faces
```

```python
#getting single face from list of faces
face= faces[0]
```

```python
x,y,w,h = face #face has four values x,y, width, height 
```

```python
# y+h and x+w for range of face
face= rgb[y:y+h, x:x+w]
```

```python
plt.imshow(face)
```

```python
#rectangle marking the face
```

```python
output= cv2.rectangle(rgb, (x, y), (x+w, y+h), (0, 0, 255), 10)
```

```python
plt.imshow(output)
```

```python
#rectangle marking all faces
```

```python
for face in faces:
    x,y,w,h = face
    output= cv2.rectangle(rgb, (x, y), (x+w, y+h), (0, 0, 255), 4)
    plt.imshow(output)

```
