```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
pil = Image.open("steven.jpg")
```

```python
arr= np.array(pil)
```

```python
#show shape of image
arr.shape
```

```python
pil
```

```python
#cut the image in in coordance to 3 dimention, 
#where first is row, second column and third is RGB 
cut = arr[:, :, 0]
```

```python
#display the image 
plt.imshow(cut)
```

```python
#creating a copy
copyImg= arr.copy()
```

```python
#here we cut the GB(1,2) color and only Red(0) is remaining ### converted layer 1 and layer 2 to 0
copyImg[:, :, [1,2]] = 0
```

```python
#display using matplot
plt.imshow(copyImg)
```

```python
### GRAY FILTER (BLACK AND WHITE)
```

```python
#multi color gray scale
#here we take mean of axis 2, which is axis of color(RGB)
grayImg= arr.mean(axis=2)
```

```python
#cmap is used to colormapping
plt.imshow(grayImg, cmap= "gray")
```
